```python
# ATM app 

import time as t

balance = 1000.00
logs = dict()
expected_pin_code = 1234

# 1. Maak een functie die de pincode van de gebruiker vraagt 
#    en checkt of deze gelijk is aan 1234. als het niet correct is, 
#    blijf om maximaal 3 keer te vragen totdat de gebruikersinvoer correct is. 

def check_pin(pin_code):
    
    # 1.1. Als de pincode niet correct is and de gebruiker heeft nog 3 pogingen,
    #      vraag dan de gebruiker om de pincode opnieuw in te voeren.
    #      Als de gebruiker nog 2 pogingen heeft, geef dan een foutmelding.
    #      Als de gebruiker nog 1 poging heeft, geef dan een foutmelding.
    #      Als de gebruiker geen pogingen meer heeft, geef dan een foutmelding en stop het programma.
    
    while True:
        max_attempts = 3
        
        if pin_code != expected_pin_code:
            max_attempts = max_attempts - 1
            warn_wrong_pin_code(max_attempts)
            if max_attempts == 0:
                warn_no_more_attempts()
                break
            else:
                pin_code = int(input("Please enter your pin-code: "))

def warn_wrong_pin_code(max_attempts):
    message = "Wrong pin-code. You have {0} more attempts.".format(max_attempts)
    print(message)
    logs[t.time_ns, message]
    
def warn_no_more_attempts():
    message = "No more attempts. The program is terminated."
    print(message)
    logs[t.time_ns, message]

# 2. Als de pincode correct is, vraag dan de gebruiker om een bedrag op te nemen.
#    Als het bedrag groter is dan de balance, geef dan een foutmelding.
#    Blijf verder om te vragen totdat juiste amount is gegeven (bvb. 250, 500, maar niet -100, 1400, enz.)

def withdraw(amount):
    while True:
        amount = float(input("Please enter the amount you would like to withdraw: "))
        if amount > balance:
            warn_cannot_withdraw_more_than_balance()
        elif amount <= 0:
            warn_cannot_withdraw_negative_amount()
        else:
            inform_withdraw_in_progress(amount)
            give_money(amount)
            inform_withdrawal_success(amount)
            break

# 3. Als het bedrag kleiner is dan de 'balance', 
# vraag dan als de geburiker wilt 'withdrawal' aanvaarden of niet.
# (op een andere manier, als de gebruiker zeker is om verder te gaan of niet)

def validate(are_you_sure, amount):
    while True:
        are_you_sure = ask_to_confirm_withdrawal(amount)
        if are_you_sure == 'y':
            return True
        elif are_you_sure == 'n':
            warn_that_withdraw_cancelled()
            return False
        else:
            warn_to_try_again()

# 4. Als de gebruiker validate bevestigt,
#    geef dan het bedrag uit en geef een melding dat het geld is uitgegeven.

def give_money(amount):
    balance = balance - amount

# 5. Als de gebruiker validate niet bevestigt,
#    geef dan een foutmelding en vraag de gebruiker om het bedrag opnieuw in te voeren.

def ask_to_try_again(amount):
    while True:
        are_you_sure = ask_to_confirm_withdrawal(amount)
        if are_you_sure == 'y':
            return True
        elif are_you_sure == 'n':
            warn_that_withdraw_cancelled()
            return False
        else:
            warn_to_try_again()


def warn_to_try_again():
    message = "Please enter a valid option to withdraw: 'y' for yes, 'n' for no."
    print(message)
    logs[t.time_ns, message]
    
def ask_to_confirm_withdrawal(amount):
    message = "Please confirm that you would like to withdraw {0}: y/n".format(amount)
    logs[t.time_ns, message]
    return input(message)
    
def warn_that_withdraw_cancelled():
    message = "Withdrawal request is cancelled."
    logs[t.time_ns, message]
    print(message)
    
def inform_withdraw_in_progress(amount):
    message = "Withdrawal of {0} is in progress..".format(amount)
    print(message)
    logs[t.time_ns, message]
    
def inform_withdrawal_success(amount):
    message = "Withdrawal of {0} is completed. Your new balance is {1}.".format(amount, balance)
    print(message)
    logs[t.time_ns, message]

def warn_cannot_withdraw_negative_amount():
    message = "Please enter a POSITIVE value (an amount grater than zero)."
    print(message)
    logs[t.time_ns, message]
    
def warn_cannot_withdraw_more_than_balance():
    message = "You can NOT withdraw more than {}.".format(balance) 
    print(message)
    logs[t.time_ns, message]
```

```python
# 6. Als de gebruiker de geldopname heeft voltooid,
#    vraag dan de gebruiker of hij/zij nog een transactie wil doen.
#   Als de gebruiker 'y' antwoordt, ga dan terug naar stap 2.
#   Als de gebruiker 'n' antwoordt, geef dan een melding dat de transactie is voltooid.

def ask_to_continue():
    while True:
        message = "Would you like to do another transaction? y/n"
        print(message)
        logs[t.time_ns, message]
        answer = input(message)
        if answer == 'y':
            return True
        elif answer == 'n':
            return False
        else:
            warn_to_try_again()
            
def inform_transaction_completed():
    message = "Transaction completed."
    print(message)
    logs[t.time_ns, message]
    
def inform_transaction_cancelled():
    message = "Transaction cancelled."
    print(message)
    logs[t.time_ns, message]
```

```python
# 9. Het uitvoeren van de functies

def print_welcome_message():
     message = "Welcome to the ATM."
     print(message)
     logs[t.time_ns, message]
     
def print_balance():
     message = "Your current balance is {}.".format(balance)
     print(message)
     logs[t.time_ns, message]

def run():
     # 9.1. Welkomstbericht
    print_welcome_message()
    
    # 9.2. Druk de balans af.
    print_balance()     
    
    # 9.3. Vraag de gebruiker om zijn/haar pincode in te voeren.
    while True:
        pin = check_pin()
        if check_pin(pin):
          withdraw()
        else:
          warn_to_try_again()
```
